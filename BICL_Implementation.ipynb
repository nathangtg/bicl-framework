```python
from google.colab import files

uploaded = files.upload()
```

## Unzip the file

### Subtask:
Provide code to unzip the uploaded file.


**Reasoning**:
Import the `zipfile` module and iterate through the uploaded files to unzip them.



```python
import zipfile

for filename in uploaded.keys():
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        zip_ref.extractall()
```
